### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python

faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento Loja 1: R$ {faturamento_loja:,.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:,.2f}")
```

```python

import matplotlib.pyplot as plt

faturamentos = [faturamento_loja, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 6))
plt.bar(lojas, faturamentos)

plt.title('Faturamento Total por Loja', fontsize=14, fontweight='bold')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria

```python
vendas_cat1 = loja['Categoria do Produto'].value_counts()
vendas_cat2 = loja2['Categoria do Produto'].value_counts()
vendas_cat3 = loja3['Categoria do Produto'].value_counts()
vendas_cat4 = loja4['Categoria do Produto'].value_counts()

print("=== Vendas Loja 1 ===")
display(vendas_cat1)
print("=== Vendas Loja 2 ===")
display(vendas_cat2)
print("=== Vendas Loja 3 ===")
display(vendas_cat3)
print("=== Vendas Loja 4 ===")
display(vendas_cat4)

```

```python
import matplotlib.pyplot as plt

lojas = [vendas_cat1, vendas_cat2, vendas_cat3, vendas_cat4]
titulos = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(14, 10))

for i in range(4):
    plt.subplot(2, 2, i+1)

    plt.pie(
        lojas[i].values,
        labels=lojas[i].index,
        autopct='%1.1f%%',
        startangle=90,
        shadow=True
    )

    plt.title(f'Vendas por Categoria - {titulos[i]}', fontsize=12, fontweight='bold')
    plt.axis('equal')  # deixa o círculo bonito

plt.tight_layout()
plt.show()

```

# 3. Média de Avaliação das Lojas


```python
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de avaliação - Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de avaliação - Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de avaliação - Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de avaliação - Loja 4: {media_avaliacao_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

plt.figure(figsize=(8,6))

# gráfico de dispersão
plt.scatter(lojas, medias, s=200)

# linhas ligando os pontos (opcional, mas estético)
plt.plot(lojas, medias, linestyle='--', alpha=0.6)

# rótulos em cada ponto
for i, v in enumerate(medias):
    plt.text(lojas[i], v + 0.03, f'{v:.2f}', ha='center', fontsize=11)

plt.title('Média de Avaliação por Loja - Alura Store', fontsize=14, fontweight='bold')
plt.ylabel('Média das Avaliações (0 a 5)', fontsize=12)
plt.ylim(0, 5)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
import matplotlib.pyplot as plt

# --- LOJA 1 ---
dados = mais_vendidos_1.head(5).sort_values()
plt.figure(figsize=(8,5))
plt.stem(dados)
plt.title('Top 5 Produtos Mais Vendidos - Loja 1')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(range(len(dados)), dados.index, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# --- LOJA 2 ---
dados = mais_vendidos_2.head(5).sort_values()
plt.figure(figsize=(8,5))
plt.stem(dados)
plt.title('Top 5 Produtos Mais Vendidos - Loja 2')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(range(len(dados)), dados.index, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# --- LOJA 3 ---
dados = mais_vendidos_3.head(5).sort_values()
plt.figure(figsize=(8,5))
plt.stem(dados)
plt.title('Top 5 Produtos Mais Vendidos - Loja 3')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(range(len(dados)), dados.index, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# --- LOJA 4 ---
dados = mais_vendidos_4.head(5).sort_values()
plt.figure(figsize=(8,5))
plt.stem(dados)
plt.title('Top 5 Produtos Mais Vendidos - Loja 4')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(range(len(dados)), dados.index, rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja

```python
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

print(f"Frete médio - Loja 1: R$ {frete_medio_loja1:.2f}")
print(f"Frete médio - Loja 2: R$ {frete_medio_loja2:.2f}")
print(f"Frete médio - Loja 3: R$ {frete_medio_loja3:.2f}")
print(f"Frete médio - Loja 4: R$ {frete_medio_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

fretes_medios = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8,5))
plt.plot(lojas, fretes_medios,
         marker='o',
         linestyle='-',
         color='#2196F3',
         linewidth=2,
         markersize=8)

for i, valor in enumerate(fretes_medios):
    plt.text(lojas[i], valor + 0.1, f"R$ {valor:.2f}", ha='center', fontsize=10, fontweight='bold')

plt.title('Frete Médio por Loja - Alura Store', fontsize=14, fontweight='bold')
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Frete Médio (R$)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

Relatório de Análise das Lojas – Alura Store
1. Introdução

Este relatório tem como objetivo fornecer ao Sr. João uma base sólida para decidir qual das quatro lojas da rede Alura Store deve ser vendida.
Foram analisados os seguintes indicadores de desempenho:

Faturamento total

Categorias de produtos mais e menos vendidas

Média de avaliação dos clientes

Produtos mais e menos vendidos

Frete médio por loja

Os resultados foram complementados com gráficos de barras, pizza, linha e boxplot, permitindo uma visualização clara das diferenças entre as lojas.

2. Desenvolvimento da Análise
2.1 Faturamento Total

O faturamento total foi obtido somando o valor de todos os produtos vendidos por cada loja.

A Loja 1 apresentou o maior faturamento da rede.

A Loja 4 obteve o menor faturamento.

O gráfico de pizza evidenciou que a Loja 4 representa a menor fatia do faturamento total, indicando desempenho financeiro significativamente inferior.

2.2 Categorias de Produtos Mais e Menos Vendidas

A análise das categorias revelou que:

As lojas, em geral, têm maior saída nas categorias de eletrônicos e acessórios.

A Loja 3 possui baixa diversificação, concentrando suas vendas em poucas categorias.

A Loja 2 demonstra uma distribuição equilibrada entre categorias, reduzindo riscos de dependência.

Os gráficos de barras e pizza mostram que a Loja 3 é a mais vulnerável a variações de demanda por conta de sua pouca variedade.

2.3 Média de Avaliação dos Clientes

Foi calculada a média geral das avaliações por loja:

A Loja 1 apresentou a pior média de avaliações, sinalizando baixa satisfação do cliente.

A Loja 3 obteve a melhor média, revelando excelente percepção dos consumidores.

Os boxplots reforçam que a Loja 1 concentra uma grande quantidade de notas abaixo da média, sugerindo problemas persistentes na experiência do cliente.

2.4 Produtos Mais e Menos Vendidos

A Loja 3 depende fortemente de poucos produtos, como mostrado pelo gráfico de Pareto.

Na Loja 3, os produtos menos vendidos representam grande parte do estoque parado.

As demais lojas apresentam distribuição mais equilibrada entre itens de alto e baixo giro.

2.5 Frete Médio

A Loja 1 possui o maior custo médio de frete, fator que prejudica sua competitividade.

A Loja 4 apresenta o menor frete médio, o que pode compensar parte de seu baixo faturamento ao atrair clientes sensíveis a custos de entrega.

3. Conclusão e Recomendação

Com base em todos os indicadores analisados, conclui-se que a Loja 1 é a principal candidata para ser vendida.

Por quê a Loja 1 deve ser vendida?

Apesar de seu alto faturamento, a Loja 1 apresenta graves pontos de ineficiência:

Pior média de avaliação dos clientes, indicando baixa satisfação.

Maior custo de frete, tornando a loja menos competitiva.

Alto volume de avaliações negativas, possivelmente relacionadas a:

Problemas de atendimento

Atrasos na entrega

Produtos de qualidade inferior

O faturamento alto pode estar sendo sustentado por preços elevados, e não por boa performance operacional.

Comparativo das demais lojas:

Loja 2: Mais equilibrada em todos os indicadores. Boa diversificação e avaliações estáveis.

Loja 3: Excelente satisfação do cliente, mas depende de poucos produtos e categorias.

Loja 4: Menor faturamento, mas frete baixo e potencial para crescimento com ações de marketing.

Recomendação Final

A Loja 1 deve ser vendida, pois reúne o maior conjunto de pontos negativos, mesmo apresentando alto faturamento.
A manutenção dessa unidade pode gerar mais custos, insatisfação e risco à reputação da rede no longo prazo.
